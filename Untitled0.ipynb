```python
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')
```

```python
fx = -5*x**2 + 10*x - 150
fx
```

```python
solve(fx)
```

```python
solve(diff(fx) > 0)
```

```python
solve(diff(fx) < 0)
```

```python
solve(diff(fx))
```

```python
plot(fx)
```
